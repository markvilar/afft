### Notebook for exporting database tables to file
This notebook dumps all the tables in database to CSV files.

```python
import pandas as pd

from pathlib import Path
from tqdm.auto import tqdm
from sqlalchemy import create_engine, inspect
```

#### Config - Set up database and export directory

```python
# Connection URI format: "postgresql://user:password@host:port/database"
DB_USER = "USERNAME"
DB_PASSWORD = "PASSWORD"
DB_HOST = "localhost"
DB_PORT = 5432
DB_NAME = "acfr_auv_messages"

EXPORT_DIR: Path = Path("/home/martin/data/acfr_telemetry_tables")
if not EXPORT_DIR.is_dir():
    raise ValueError(f"directory does not exist: {EXPORT_DIR}")
```

#### Create database connection and inspect tables

```python
DATABASE_URI = f"postgresql+psycopg://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
engine = create_engine(DATABASE_URI)

# Create an inspector object
inspector = inspect(engine)

# Get a list of table names
table_names = inspector.get_table_names()
```

#### Read database tables and export to CSV files

```python
for table_name in tqdm(table_names):
    df: pd.DataFrame = pd.read_sql_table(table_name, con=engine)

    # Write to a CSV file named 'output.csv'
    df.to_csv(EXPORT_DIR / f"{table_name}.csv")
```
