### Get sea level from the Stormglass API using the Metocean package

```python
import dataclasses

from pathlib import Path

import pandas as pd


import afft.metocean as metocean


START_DATE: str = "20090101"
END_DATE: str = "20211231"


EXPORT_DIRECTORY: Path = Path("/data/exos_01/metocean_sea_level_hourly")


@dataclasses.dataclass
class SeaLevelRequestContext:
    """Class representing a site irradiance request."""

    label: str
    request: metocean.SeaLevelRequest


reference_sites: list[tuple] = [
    ("qd61g27j", 115.45961, -32.03382, START_DATE, END_DATE),
    ("qdc5ghs3", 114.04409, -28.84709, START_DATE, END_DATE),
    ("qdch0ftq", 113.94659, -28.81328, START_DATE, END_DATE),
    ("qdchdmy1", 114.01354, -28.70839, START_DATE, END_DATE),
    ("qtqxshxs", 121.82688, -14.12490, START_DATE, END_DATE),
    ("r7jjskxq", 153.47815, -27.13237, START_DATE, END_DATE),
    ("r7jjss8n", 153.47918, -27.13237, START_DATE, END_DATE),
    ("r7jjssbh", 153.47918, -27.13134, START_DATE, END_DATE),
    ("r23m7ms0", 148.15630, -42.64060, START_DATE, END_DATE),
    ("r29mrd5h", 148.34169, -41.25289, START_DATE, END_DATE),
    ("r29mrd12", 148.33929, -41.25358, START_DATE, END_DATE),
    ("r234xgje", 148.00455, -43.13722, START_DATE, END_DATE),
    ("r23685bc", 148.00901, -43.13344, START_DATE, END_DATE),
]


contexts: list = list()
for label, longitude, latitude, start_date, end_date in reference_sites:
    request: metocean.SeaLevelRequest = metocean.SeaLevelRequest(
        longitude=longitude,
        latitude=latitude,
        start_date=start_date,
        end_date=end_date,
        api=metocean.SeaLevelAPI.WORLDTIDES,
    )
    contexts.append(SeaLevelRequestContext(label=label, request=request))


# Fetch sea level data from the WorldTides API
for context in contexts:
    df: pd.DataFrame = metocean.request_sea_level(context.request)
    filename: str = f"{context.label}_{context.request.start_date}_{context.request.end_date}_sea_level.csv"
    export_path: Path = EXPORT_DIRECTORY / filename
    df.to_csv(export_path, index=False)
```
