### Notebook for visualizing pressure sensor readings
This notebook plots series of pressure sensor readings, sea levels, and tide-corrected pressure sensor readings.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from pathlib import Path
from loguru import logger
```

```python
readings_file: Path = Path("/data/exos_01/acfr_messages_v3_processed/qd61g27j_20100421_022145_pressure_parosci.csv")
if not readings_file.is_file():
    raise FileNotFoundError(f"missing file: {readings_file}")

readings_frame: pd.DataFrame = pd.read_csv(readings_file, parse_dates=["timestamp"])
logger.info(readings_frame)
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1, sharex=True)

sns.scatterplot(
    ax=axes[0],
    data=readings_frame,
    x="timestamp",
    y="depth",
    s=1,
)

sns.scatterplot(
    ax=axes[0],
    data=readings_frame,
    x="timestamp",
    y="corrected_depth",
    s=1,
)

sns.scatterplot(
    ax=axes[1],
    data=readings_frame,
    x="timestamp",
    y="sea_level",
    s=1,
)

figure.suptitle("Tide-Corrected Pressure Sensor Readings - Rottnest 20100421", fontsize=10)

axes[0].yaxis.set_inverted(True)
axes[0].set_xlabel("Timestamp")
axes[0].set_ylabel("Depth [m]")

axes[1].set_xlabel("Timestamp")
axes[1].set_ylabel("Tide [m]")

figure.tight_layout()
```
